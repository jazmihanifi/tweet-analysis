```python
import re
```

```python
tweet = data['Tweet']
text_tweet = []
for text in tweet:
  text = text.lower()
  text = re.sub("(user)","",text)
  text = re.sub("(rt)","",text)
  text = re.sub("(url)","",text)
  text = re.sub(r"https://t.co/\w+"," ",text)
  text = re.sub(r"\\n","",text)
  text = re.sub("\n","",text)
  text = re.sub(r"\\x.."," ",text)
  text = re.sub(r"#\w+","",text)
  text = re.sub("\W"," ",text)
  text = re.sub("( ){2,15}"," ",text)
  text_tweet.append(text)
  

text_tweet
```

```python
#mengecek hasil
print(data['Tweet'][79])
print(text_tweet[79])

print(data['Tweet'][11])
print(text_tweet[11])
```
